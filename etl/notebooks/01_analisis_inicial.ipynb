# Análisis inicial de Bank Marketing
Este notebook acompaña al pipeline modular. La ejecución oficial se realiza con `python -m etl.pipeline`.


```python
import pandas as pd
df = pd.read_csv('../../data/raw/bank.csv')
df.head()
```

```python
print('Filas y columnas:', df.shape)
display(df.dtypes)
display(df.describe(include='all').T)
```

```python
categorical = df.select_dtypes(include='object').columns
unknown_counts = (df[categorical] == 'unknown').sum().sort_values(ascending=False)
display(unknown_counts[unknown_counts > 0])
```

```python
conversion_rate = (df['y'].eq('yes').mean() * 100)
print(f'Tasa de conversión: {conversion_rate:.2f}%')
```
